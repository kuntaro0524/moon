```python
import numpy as np
import matplotlib.pyplot as plt
 
# 円グラフを描画
x = np.array([10]*36)
plt.pie(x, wedgeprops={'linewidth': 1, 'edgecolor':"white"})

plt.savefig("small_wedge.png", format="png", dpi=300)
```

```python
label = ["Apple", "Banana"]
colors = ["lightpink", "none"]
x = np.array([50,50])
plt.pie(x, startangle=-90, colors=colors)

plt.savefig("300_dpi_scatter.png", format="png", dpi=300, wedgeprops={'linewidth': 3, 'edgecolor':"white"})

# plt.pie(x, labels=label, counterclock=False, startangle=90, colors=colors,  wedgeprops={'linewidth': 3, 'edgecolor':"white"}, explode=[0.2, 0],))
```
