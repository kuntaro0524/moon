# Sample nameでグループを作成する

```python
grouped = df_done.groupby("sample_name")
grouped.size()
```

# groupごとに処理をしてみるよ

```python
for name, group in grouped:
    # selection
    for puck, pinid in zip(group.puckid, group.pinid):
        print(name, puck,pinid)
    print("#####################")
    
```
